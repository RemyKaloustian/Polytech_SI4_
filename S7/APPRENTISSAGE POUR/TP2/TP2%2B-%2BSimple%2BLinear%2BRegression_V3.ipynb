```python
import numpy as np                          #imports library numpy as np
import matplotlib.pyplot as plt             #imports pyplot as plt
from sklearn import linear_model
from scipy import stats


#Question 1
n=500
x = np.linspace(1.0, 100.0,n) #Distributes the points 
x = x.reshape(-1,1) #is necessary for the linear regression to work. Reshapes x in a right vector form.

#Question 2
epsilon = np.random.randn(n,1) * np.sqrt(104.04) #Our epsilon parameter
y = 2*x + 25.2 + epsilon #line equation

#Question 3
plt.scatter(x,y)
plt.title('Random cloud with residuals in red')  #gives a title to the graph
plt.xlabel('x')            #gives the y axis a label
plt.ylabel('y')            #gives the y axis a label

#Question 4
regr = linear_model.LinearRegression() #olves the linear regression problem
regr.fit(x,y) #gets the interesting parameters of the regression 
y_hat = regr.predict(x) # the estimate of Y

#Question 5
plt.plot(x,y_hat,'y')
plt.plot(x, y-y_hat,'r')
plt.show()

#Question 6
sig_x_carre = sum((x-np.mean(x))**2) #Used to simplify the calculation
h = 1/n +((x-np.mean(x))**2)/sig_x_carre #Used to simplify the calculation
sig_carre_hat = 1/(n-2)*sum((y-y_hat)**2) #Used to simplify the calculation

#Question 7
t = (y-y_hat)/np.sqrt(sig_carre_hat*(1-h)) #Making the estimate of the residuals
plt.plot(x,t,'b')
plt.title('The standardized residuals')  #gives a title to the graph
plt.xlabel('x')            #gives the y axis a label
plt.ylabel('y')  

plt.show()

#Question 8 
print('The coefficient of determination is : ' , regr.coef_)

#Question 9 
#talpha2 = stats.t.ppf(1-/2, n-2)
#print("talpha is : ", talpha2)



```
