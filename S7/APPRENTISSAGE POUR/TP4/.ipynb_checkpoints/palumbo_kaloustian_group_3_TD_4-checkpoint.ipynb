#### Question 3,  Pour step = 2, et $x^{(0)}$ = -50

```python
k = 0
N=20
step = 2
xk = []
fxk = []
ek = []

xk.append(-50)
while k < N:
    fxk_t = (1/5)*xk[k]-8
    xk_t = xk[k] + step*(-fxk_t)
    
    ek_t = ((1/10)*xk[k]**2 -8*xk[k] + 10) - ((1/10)*40**2 - 8*40 + 10)
    xk.append(xk_t)
    fxk.append(fxk_t)
    ek.append(ek_t)
    k += 1
```

####  Pour step = 12, et $x^{(0)}$ = -50

```python
k = 0
N=20
step = 12
xk = []
fxk = []
ek = []
xk.append(-50)
while k < N:
    fxk_t = (1/5)*xk[k]-8
    xk_t = xk[k] + step*(-fxk_t)
    
    ek_t = ((1/10)*xk[k]**2 -8*xk[k] + 10) - ((1/10)*40**2 - 8*40 + 10)
    xk.append(xk_t)
    fxk.append(fxk_t)
    ek.append(ek_t)
    k += 1

plt.subplot(2,1,1)
plt.plot(fxk, 'r')

plt.subplot(2,1,2)
plt.plot(ek, 'r')
plt.show()

plt.plot(xi, y, 'r')
xk_array = np.array(xk)
yk = (1/10)*xk_array**2 - 8*xk_array + 10
plt.plot(xk, yk, 'yo');
```

On constate plusieurs impacts du pas sur l'algorithme de convergence. Il y a tout d'abord un impact sur la zone de recherche. On voit que plus le pas est grand, plus la distance entre les deux points de deux iterations successives deviendra grande. On en déduit que pour le pas, l'utilisation d'une bonne valeur est très importante.
Par exemple, dans le cas où l'on a un minimum local et un minimum global différents,  le choix d'un pas réduit pourrait limiter la recherche autour d'un minimum local. Au contraire, l'utilisation d'un pas plus grand peut permettre de trouver le minimum global et non local. On élargit alors notre zone de recherche.
Cependant, comme on peut le voir si on utilise un pas élevé (step = 12), l'algorithme peut devenir moins efficace et s'éloigner du minimum au lieu de s'en rapprocher. Le but est donc de bien jauger le pas afin de trouver le minimum global.
