```python
# -*- coding: utf-8 -*-
"""
Created on Fri Sep 30 10:52:39 2016

@author: RÃ©my Kaloustian
"""

#this code creates two figures and displays them

#%matplotlib inline -> Makes the figures appear in the notebook

import numpy as np #import for maths tools
import matplotlib.pyplot as plt #import figures creation
x1 = np.linspace(0.0,5.0) #divides the [0,5] interval in 50 values in a linear vector
x2  = np.linspace(0.0,2.0)#divides the [0,2] interval in 50 values in a linear vector
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)#sames as below, just multiplying by exp to attenuate the sinusoid
y2 = np.cos(2 * np.pi  * x2) # y2 is now a vector, the cos is  applied to each element of the vector
plt.subplot(2,1,1) #Create the first figure, 2 lines, one column, in the first block
plt.plot(x1,y1,'yo-')#set the axis, y for yellow, o for point in o shape, -  for continuous line linking the points
plt.title('A table of 2 subplots')#set the general title
plt.ylabel('Damped oscillation')#Set the name of the y axis 
plt.subplot(2,1,2)#Trace the second figure, 2 lines, one column, second block
plt.plot(x2,y2, 'r.-') #Set the axis, r for red, . for little points shape, - (same as higher)
plt.xlabel('time (s)') #Set the title of the x axis
plt.ylabel('Undamped') #Set the title of the y axis
plt.show() #Displays the figures
```

```python
#This program finds teh highest eignevalue of a matrix

import numpy as np
from scipy import linalg as LA
A = np.array([[2,0,0], [0,3,4], [0,4,9]])#matrix 3*3
xk = np.array([[0.1], [0.1], [0.1]])#vector to initialize loop
for k  in range (1,100):#from 1 to 100
    yk = A.dot(xk)#yk = A*x
    xk = yk/LA.norm(yk, np.inf)#calculates the norm infinite of yk
    
#We got the highest eigenvalue of A
x = xk/LA.norm(xk,2)#calculates the norm 2 of xk
value = (x.transpose()).dot(A.dot(x)) #value is xT * A
print(value)#prints 11
```
